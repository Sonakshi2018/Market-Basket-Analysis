As we are doing we need row wise value description of various customers of what they purchased and since we have data in a different format, we will create Pivot table.

```python
Pivot = FirstVisit_Sales.pivot_table(index=['ClientID'], columns=['Item'], aggfunc=[len], fill_value=0)
```

```python
pd.set_option('display.max_columns', None)
Pivot.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

item_sets = Pivot.applymap(encode_units)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(item_sets, min_support=0.01, use_colnames=True)
```

```python
frequent_itemsets.sort(['support'],ascending = [False])
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules = rules.sort_values(by='lift',ascending = False)
rules
```

```python
rules[ (rules['lift'] >= 2) &
       (rules['confidence'] >= 0.5) ]
```

```python
We can see that customers paying for Weekend Private are also paying Payment on Account.
Lot of customers are purchasing 1 Child Membership Plan.
There are customers who are paying for Halloween Party and then paying for 1 Child Membership Plan.
```
